这是hot100的第35题：LRU 缓存
请你设计并实现一个满足  LRU (最近最少使用) 缓存 约束的数据结构。
实现 LRUCache 类：
LRUCache(int capacity) 以 正整数 作为容量 capacity 初始化 LRU 缓存
int get(int key) 如果关键字 key 存在于缓存中，则返回关键字的值，否则返回 -1 。
void put(int key, int value) 如果关键字 key 已经存在，则变更其数据值 value ；如果不存在，则向缓存中插入该组 key-value 。如果插入操作导致关键字数量超过 capacity ，则应该 逐出 最久未使用的关键字。
函数 get 和 put 必须以 O(1) 的平均时间复杂度运行。

```python
from collections import defaultdict

class LRUCache:

    def __init__(self, capacity: int):
        self.capacity = capacity
        self.dic = defaultdict()
        self.length = 0

    def get(self, key: int) -> int:
        if key in self.dic:
            v = self.dic[key]
            del self.dic[key]
            self.dic[key] = v
            return self.dic[key]
        else:
            return -1

    def put(self, key: int, value: int) -> None:
        if  key in self.dic:
            del self.dic[key]
            self.dic[key] = value
        elif self.length < self.capacity and key not in self.dic:
            self.dic[key] = value
            self.length += 1
        else:
            first_key = next(iter(self.dic))
            self.dic.pop(first_key)
            self.dic[key] = value



# Your LRUCache object will be instantiated and called as such:
# obj = LRUCache(capacity)
# param_1 = obj.get(key)
# obj.put(key,value)
```

这里用到的是python3.7之后的字典的有序性，新加入后，新键值对在末尾，第一个也就是最少使用的值，get用字典的哈希查找就行，查找后，更新到末尾，put就用了length进行比较，每次插入就更新长度，要是键在其中，就不需要判断，把原有的删除，在最后添加即可，当然要是没在，就要判断一下长度关系了，长了就要把第一个删掉，没有长就不需要删除第一个

学到的地方：
1.字典现在具有有序性，后添加的回在字典后面
2.如何获得字典等非列表元素的第一个值： iter(dic)用来把字典的键化为迭代器，当然他不储存 next(iter())用来获取迭代器的第一个值，也就是字典的第一个键

```python
class node:
    def __init__(self,key=0,val=0,pre=None,next=None):
        self.key = key
        self.val = val
        self.pre = pre
        self.next = next
        
class LRUCache:
    def __init__(self, capacity: int):
        self.capacity = capacity
        self.length = 0
        self.end = self.head = node()
        self.cache = {}

    def get(self, key: int) -> int:
        if key in self.cache and self.cache[key] == self.end:
            return self.end.val
        elif key in self.cache and self.cache[key] != self.end:
            m = self.cache[key]
            l = m.pre
            n = m.next
            l.next = n
            n.pre = l
            m.next = None
            self.end.next = m
            m.pre = self.end
            self.end = m
            return m.val
        else:
            return -1

    def put(self, key: int, value: int) -> None:
        if key in self.cache and self.cache[key] == self.end:
            self.end.val = value
        elif key in self.cache and self.cache[key] != self.end:
            m = self.cache[key]
            l = m.pre
            n = m.next
            l.next = n
            n.pre = l
            m.val =value
            m.next = None
            m.pre = self.end
            self.end.next = m
            self.end = m
        else:
            if self.length < self.capacity:
                m = node(key=key,val=value)
                self.cache[key] = m
                m.pre,self.end.next = self.end,m
                self.end = self.end.next
                self.length += 1
            else:
                n = self.head.next.key
                self.cache.pop(n)
                self.head = self.head.next
                m = node(key=key,val=value)
                self.cache[key] = m
                m.pre,self.end.next = self.end,m
                self.end = m

# Your LRUCache object will be instantiated and called as such:
# obj = LRUCache(capacity)
# param_1 = obj.get(key)
# obj.put(key,value) 
```

如果不用defaultdict话，就要使用哈希表降低查询的时间，双向链表用来把put和get更新的内容进行更新
要注意的是，在key对应的节点是最后一个节点时，它没有后续节点，所以要把它单独列出来，不用和后面的节点断开，后面的节点是none，也没有节点了
