这是hot100的第28题 两数相加 给你两个 非空 的链表，表示两个非负的整数。它们每位数字都是按照 逆序 的方式存储的，并且每个节点只能存储 一位 数字。

请你将两个数相加，并以相同形式返回一个表示和的链表。

你可以假设除了数字 0 之外，这两个数都不会以 0 开头。



```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def addTwoNumbers(self, l1: Optional[ListNode], l2: Optional[ListNode]) -> Optional[ListNode]:
        s = ListNode()
        ans = s
        dig = 0
        while l1.next and l2.next:
            m = dig
            dig = (l1.val+l2.val+dig)//10
            l1.val = (l1.val+l2.val+m)%10
            s.next = l1
            l1 = l1.next
            l2 = l2.next
            s = s.next
            s.next = None
        if l1.next and not l2.next:
            m = dig
            dig = (l1.val+l2.val+dig)//10
            l1.val = (l1.val+l2.val+m)%10
            s.next = l1
            s = s.next
            while s:
                if s.next:
                    s = s.next
                    m = dig
                    dig = (s.val+dig)//10
                    s.val = (s.val+m)%10
                elif not s.next:
                    if dig == 1:
                        s.next = ListNode(1)
                    break
        elif l2.next and not l1.next:
            m = dig
            dig = (l1.val+l2.val+dig)//10
            l2.val = (l1.val+l2.val+m)%10
            s.next = l2
            s = s.next
            while s:
                if s.next:
                    s = s.next
                    m = dig
                    dig = (s.val+dig)//10
                    s.val = (s.val+m)%10
                elif not s.next:
                    if dig == 1:
                        s.next = ListNode(1)
                    break
        elif not l1.next and not l2.next:
            m = dig
            dig = (l1.val+l2.val+dig)//10
            l1.val = (l1.val+l2.val+m)%10
            s.next = l1
            s = s.next
            if dig == 1:
                s.next = ListNode(1)
        return ans.next
```

这个思路不复杂，就是链表在处理的时候会晕，不知道处理到哪一节了
首先链表是尾部对其的，当两个数字都没到最高位是，一样的方法，将两者相加，储存是否需要进位，用l1储存起来计算后的值放到s，然后l1，l2同时向后推进
当有一个达到最高位了，两者相加，储存是否进位，同时把s的next指向没有到最高位的l，再继续计算，后面只需要计算有无进位就可以了，当计算到最后一个的最高位后，如果需要进位，就加一个listnode（1）
当两者同时达到最高位，两者相加，储存是否进位，如果是的话，就s的next加一个listnode（1），此时s还是l1的最高位

这个是迭代的方法

这个写法太复杂了，有简化的方法
1.全部放在内部,而且也没有必要把l1作为储存
2.为了防止在有一个到达最高位另一个没有后，可以用中间变量储存值，当为none时是0，除此之外就是节点的值
3.在next的时候没有了就判断一下

学到的地方：
1.n = x if x else y用来简化判断

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def addTwoNumbers(self, l1: Optional[ListNode], l2: Optional[ListNode]) -> Optional[ListNode]:
        dummy = ListNode()
        cur = dummy
        dig = 0
        while l1 or l2 or dig:
            v1 = l1.val if l1 else  0
            v2 = l2.val if l2 else  0
            total = (v1+v2+dig)%10
            dig = (v1+v2+dig)//10
            cur.next = ListNode(total)
            cur = cur.next
            l1 = l1.next if l1 else None
            l2 = l2.next if l2 else None
        return dummy.next
```

下面写一个递归的方法：
这个方法也是规避了使用复杂判断条件的，用中间变量，两次累加

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def addTwoNumbers(self, l1: Optional[ListNode], l2: Optional[ListNode]) -> Optional[ListNode]:
        def ad(l1,l2,carry):
            if not l1 and not l2 and not carry:
                return None
            s = carry
            if l1:
                s += l1.val
                l1 = l1.next
            if l2:
                s += l2.val
                l2 = l2.next
            carry = s//10
            return ListNode(s%10,ad(l1,l2,carry))
        cur = ListNode(0,ad(l1,l2,0))
        return cur.next
```
